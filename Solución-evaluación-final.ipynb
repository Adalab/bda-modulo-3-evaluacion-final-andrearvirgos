```python
def duplicados_columnas(df):
    total_filas = len(df)
    resultados = {}

    for columna in df.columns:
        conteo_repeticiones = df[columna].duplicated().sum()
        conteo_unicos = total_filas - conteo_repeticiones
        porcentaje_rep = (conteo_repeticiones / total_filas) * 100

        resultados[columna] = {
            'V-únicos': conteo_unicos,
            'V-repetidos': conteo_repeticiones,
            'Reps.porcentaje': f'{porcentaje_rep:.2f}%'
        }
    return pd.DataFrame(resultados).T

resultados_conteos = duplicados_columnas(df_history)
print(resultados_conteos)
```

## Unión datasets:

### Primero, limpieza de los datasets pre-mergeo, por eficiencia:

```python
# 1. Gestión registro anómalo, salario negativo. Convertir a nulo, para agruparlo con el resto de nulos de la columna y gestionarlos a posteriori juntos.
df_history.loc[df_history['Salary'] <=0, 'Salary'] = np.nan
```

```python
# verificación de la conversión a nulo
print(f"El valor mínimo de Salary ahora es: {df_history['Salary'].min():,.2f}")
```

```python
print(f"Ahora el conteo de valores nulos en Salary es: {df_history['Salary'].isnull().sum()}")

# dado que antes este recuento sumaba 4238, la conversión ha funcionado! 
```

```python
# 2. Creación de una columna binaria, por asignación directa, que permita eliminar las columnas de Year y Month que tienen unos tantos por cientos elevadísimos de nulos.
# Sustituir esas dos columnas que no aportan información de calidad (es información escasa e irrelevante) por una sola columna que conteste realmente a la pregunta de si la suscripción del cliente está activa o no.
# Ésto redunda en la eficiencia y rapidez del proceso de unión.

df_history['Status'] = df_history['Cancellation Year'].notna().astype(int)  # uso la columna cancellation year, en vez de la de month porque es más fiable (tiene 12 valores únicos, no como la de cancellation month que tiene 13: 12 meses+ NaN)
```

```python
df_history['Status'].value_counts() # Comprobación que columna nueva tiene valores 0 = No cancelado/ 1= Cancelado)
```

```python
# Eliminación columnas innecesarias (estas 2 columnas pasan a ser ruido visual, porque ya tenemos el estado de la suscripción en 'Status')
df_history.drop(columns=['Cancellation Year', 'Cancellation Month'], inplace=True)
```

```python
# codificación numérica para la variable categórica 'Education' (mapeo de menor a mayor):

education_map = {
    'High School or Below': 1,
    'College': 2,
    'Bachelor': 3,
    'Master': 4,
    'Doctor': 5
}
df_fusionado['Education_Codes'] = df_fusionado['Education'].replace(education_map).astype(int)
df_fusionado.drop(columns=['Education'], inplace=True)
```

```python
df_fusionado['Education_Codes'].value_counts()
```

```python
marital_map = {
    'Single': 0,
    'Married': 1,
    'Divorced': 2
}
df_fusionado['Marital_Status_Codes'] = df_fusionado['Marital Status'].replace(marital_map).astype(int)
df_fusionado.drop(columns=['Marital Status'], inplace=True)
print("Columna 'Marital Status' codificada")
```

```python
loyalty_map = {
    'Aurora': 0,
    'Nova': 1,
    'Star': 2
}
df_fusionado['Loyalty_Card_Codes'] = df_fusionado['Loyalty Card'].replace(loyalty_map).astype(int)
df_fusionado.drop(columns=['Loyalty Card'], inplace=True)
print("Columna 'Loyalty Card' cod!")
```

## Visualizaciones: 

```python
def archivar_grafica(nombre):
    ruta = f"images/{nombre}.png"
    plt.savefig(ruta, dpi=300, bbox_inches="tight")
    print(f" Gráfica archivada en: {ruta}")
```

```python
# 1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

# etiquetas de meses
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dec']

#  Agrupación datos
vuelos_mes = df_fusionado.groupby('Month')['Flights Booked'].sum().reset_index()

# etiquetas en el df
# acceso a la lista x el número de mes (columna 'Month') menos 1, por conteo en pythonés

vuelos_mes['Month_Label'] = vuelos_mes['Month'].apply(lambda x: meses[x - 1])

#gráfica
plt.figure(figsize=(8, 4))
# columna de etiquetas 'Month_Label' en el eje X
sns.barplot(x='Month_Label', y='Flights Booked', data=vuelos_mes) 

plt.title('1. Vuelos reservados por mes')
plt.xlabel('Mes')
plt.ylabel('Total de vuelos reservados')

plt.xticks(rotation=45) 
archivar_grafica('1. Vuelos reservados por mes')
plt.show()
```

##### El año comienza con un nivel bajjo y estable entre Enero y Febrero. Hay un crecimiento desde Marzo (con una leve caída en Abril) hasta Julio, y en Agosto empiezan a disminuir hasta Noviembre. En Diciembre vuelven a experimentar un crecimiento. Podría concluirse que en periodos convencionales de vacaciones, las reservas aumentan. 

```python
# 2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente?
# uso df completo, en vez de una muestra
plt.figure(figsize=(8, 4))

sns.scatterplot(x='Distance', y='Points Accumulated', data=df_fusionado, alpha=0.6, color='lightgreen') 

plt.title('2. Distancia vuelos vs. puntos acumulados')
plt.xlabel('Distancia recorrida')
plt.ylabel('Puntos Acumulados')
plt.grid(True, linestyle=':')

archivar_grafica("2. Distancia vs puntos acumulados") 
plt.show()
```

##### En este gráfico de dispersión se ve una correlación positiva muy clara: A mayor distancia recorrida, mayor acumulación de puntos (lo esperable puesto que en general, los programas de fidelización recompensan en función del gasto que implica muchas veces la distancia).

```python
# 3. ¿Cuál es la distribución de los clientes por provincia o estado?
conteo_provincia = df_fusionado['Province'].value_counts()

# 2. Visualización
plt.figure(figsize=(5, 10)) 
# argumento 'order' con provincias ordenadas de mahyor a menor por la función value_counts()
sns.countplot(y='Province', 
              data=df_fusionado, 
              order=conteo_provincia.index,
              color='pink') 

plt.title('3. Distribución clientela por provincia')
plt.xlabel('Número de clientes')
plt.ylabel('Provincia')
plt.grid(axis='x', linestyle='--')

archivar_grafica("3. Distribucion clientela por Provincia") 

plt.show()
```

##### La gráfica muestra la distribución de la clientela base de la aerolínea, destacando una concentración muy fuerte en las provincias más pobladas de Canadá (Ontario, British Columbia y Quebec). Hay una caída significativa en el volumen de clientes en la cuarta provincia, Alberta (es poco más de un cuarto del volumen de Quebec). La estrategia de negocio de la aerolínea está visiblmente ligada a los grandes centros urbanos. 

```python
# 4.  ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?
# Niveles educativos agrupados por códigos de columna
salario_vs_educacion = df_fusionado.groupby('Education_Codes')['Salary'].mean().reset_index()

# asignación etiquetas para legibilidad
etiquetas_edu = {1: 'High School', 2: 'College', 3: 'Bachelor', 4: 'Master', 5: 'Doctor'}
salario_vs_educacion['Nivel_formativo'] = salario_vs_educacion['Education_Codes'].map(etiquetas_edu)

plt.figure(figsize=(10, 6))

sns.barplot(x='Nivel_formativo', y='Salary', data=salario_vs_educacion, color='orange') 

plt.title('4. Salario promedio por nivel formativo')
plt.xlabel('Nivel formativo')
plt.ylabel('Salario promedio')
plt.xticks(rotation=45)

archivar_grafica('4. Salario promedio por nivel formativo') 
plt.show()
```

##### El salario tiende a ser más elevado en función del nivel formativo. Hay una anomalía en las categorías 'College' y 'Bachelor' que podrían sugerir que, los títulos profesionales de College tienen una rentabilidad salarial promedio ligeramente superior a los títulos generales de Bachelor. Podría deberse a que las formaciones técnicas, especializadas tienden a tener una salida laboral más rápida que las licenciaturas que pueden requerir de más formación (máster, doctorado), para alcanzar sueldos altos. O puede tener que ver con la propia ideosincrasia laboral de Canadá.

```python
# 5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad? 

conteo_tarjeta = df_fusionado['Loyalty_Card_Codes'].value_counts()

etiquetas_tarjeta = {2: 'Star', 1: 'Nova', 0: 'Aurora'}

labels_tarjeta = [etiquetas_tarjeta[code] for code in conteo_tarjeta.index]
sizes_tarjeta = conteo_tarjeta.values

plt.figure(figsize=(5, 5))

# autopct para mostrar porcentaje y colors para diferenciar porciones
colores_magma = sns.color_palette('magma')[::-1]
plt.pie(sizes_tarjeta, labels=labels_tarjeta, autopct='%1.1f%%', startangle=90, colors=colores_magma)

plt.title('5. Proporción de clientes por tipo de tarjeta de fidelidad')
plt.axis('equal') # Asegura que el pastel sea un círculo perfecto

archivar_grafica('5. Proporción de clientes por tipo de tarjeta de fidelidad')
plt.show()
```

##### El nivel Star (la tarjeta de mayor valor/beneficios) domina la base de clientes, representando el 45.4% de todos los miembros. Casi la mitad de la clientela está en el nivel de mayor gasto/viajes frecuentes. La prioridad debe ser la retención de los clientes Star, ya que cualquier pérdida en este segmento tiene el mayor impacto negativo en los ingresos de la compañía.

```python
# 6. ¿Cómo se distribuyen los clientes según su estado civil y género?
# Renombramos las columnas codificadas para que las etiquetas sean claras en el gráfico
df_fusionado['Estado_civil'] = df_fusionado['Marital_Status_Codes'].map({0: 'Solteras', 1: 'Casadas', 2: 'Divorciadas'})
df_fusionado['Géneros'] = df_fusionado['Gender_Encoded'].map({0: 'Hombre', 1: 'Mujer'})

# 2. Visualización
plt.figure(figsize=(10, 6))
# 'Estado_civil' en el eje X, barras agrupadas por 'Género'
sns.countplot(x='Estado_civil', hue='Géneros', data=df_fusionado, palette='Greens') 

plt.title('6. Distribución de clientes por estado civil y géneros')
plt.xlabel('Estado civil')
plt.ylabel('Número de clientes')

archivar_grafica('6. Distribución de clientes por estado civil y género')
plt.show()
```

##### El estado civil de las personas casadas es el más numeroso. Y los géneros (los únicos dos que se contenplan) están muy a la par en todos los estados civiles posibles. Con lo que se puede deducir que las estrategias de retención/fidelización no tienen que contemplar tanto el género como el estado civil, enfocándose en el grupo mayoritario de personas casadas.
